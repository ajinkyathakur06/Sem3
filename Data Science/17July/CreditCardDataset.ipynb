```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("D:/Sem 3/Data Science/BankChurners.csv")
```

```python
df['Attrition_Flag'].value_counts()
```

```python
#print count of Attrited Customers from Attrition_Flag column
print("Count of Attrited Customers:", df['Attrition_Flag'].value_counts()['Attrited Customer'])
```

```python
print("Count of Attrited Customers:", df[df['Attrition_Flag'] == 'Attrited Customer'].shape[0])
```

```python
#delete rows where Education_Level is 'Unknown'
df = df[df['Education_Level'] != 'Unknown']
```

```python
#group by Income_Category 
income_group = df.groupby('Income_Category')['Attrition_Flag'].value_counts().unstack().fillna(0)
#plot the count of Attrited Customers by Income_Category
income_group.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Count of Attrited Customers by Income Category')
plt.xlabel('Income Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Attrition Flag')
plt.show()
```
